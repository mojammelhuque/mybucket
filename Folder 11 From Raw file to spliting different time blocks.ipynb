## Dual Axis Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')
plt.figure(figsize = (25,8))

fig = plt.figure()
ax = fig.add_subplot(111)
ax1= sns.scatterplot(data=df, x='Time', y ='Average Rotary Speed rpm',color ='red', label = 'RPM')

ax2 = ax.twinx()

ax2= sns.scatterplot(data=df, x='Time', y ='Rate of Penetration m/h',color ='green',label = 'ROP')
ax.legend(loc =0)

ax.set_xlabel("Time")
ax1.set_ylabel("Average Rotary Speed rpm")
ax2.set_ylabel("Rate of Penetration m/h")
ax2.set_ylim(0, 35)
ax.set_ylim(0,500)
plt.show()


```

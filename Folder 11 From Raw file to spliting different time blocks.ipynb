## Seperating rows by time block 

## Time block 1 : 17-Sep-2007: 5:00AM to 5:00 PM

```python
df_time_block1 = df[(df["Time"] > '2007-09-17 05:00:00.000') & (df["Time"] < '2007-09-17 17:00:00.00')]
df_time_block1.describe()
```

```python
df_time_block1.head(5)
```

```python
plt.figure(figsize = (30,10))
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale = 2)

ax= sns.scatterplot(data=df_time_block1, x='Time', y ='Hole_Depth_(MD)_m', color = 'blue', alpha = 1)

ax.invert_yaxis()
plt.title('Hole Depth(m) vs Drilling Time for Block 1',fontsize = 20)
plt.legend(labels=["Hole_Depth_(MD)_m"])

plt.show()
```

### Saving seperated data frames (Time Block 1) into CSV file
